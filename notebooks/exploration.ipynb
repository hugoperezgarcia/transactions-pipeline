## Conslusion

### Descripcion del dataset

Dataset de transacciones financieras con 284.807 registros y 31 columnas.
No contiene timestamps explicitos, solo una variable Time relativa.

### Problemas
* No hay una fecha real
* Dataset muy desbalanceado (Class) 99.82% Vs. 0.17%
* Hay 1081 registros duplicados
    * Los duplicados aparecen en filas consecutivas, por lo que pueden ser transacciones que esten durando mas de 1 segundo, o que se hagan justo terminando un segundo y salga en ambos.

### Decisiones
* No disponemos de un identificador de transacción ni de semántica temporal más fina, por lo que se asume que filas completamente idénticas representan duplicados técnicos.
* Se creara columna time_ts a partir de la columna Time
* Se metera ingestion_date para particionado
* Formato de salida: Parquet
